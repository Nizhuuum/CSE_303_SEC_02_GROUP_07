```python
# Function to get question number

def get_question_no(file, column_number):
    to_list = file.values.tolist()                      # file is the full data which we loaded first
    question_no = to_list[1:][0][column_number]         # can be write as funtion
    print("Question number is: ",question_no)
    return question_no
```

```python
mid_question_3 = get_question_no(file, 7)   # q3 is on column 7
print(mid_question_3)
#print(type(mid_question_3))
```

```python
# Function to modify column values for evaluation

def modify(eval_id, evaluate_data, question_no, exam):
    evaluate_data_l = evaluate_data.values.tolist()
    #print("Before: ", evaluate_data_l[0])
    for i in range(len(evaluate_data_l)):
        evaluate_data_l[i][0] = eval_id + i
        evaluate_data_l[i][1] = 100 + i
        evaluate_data_l[i][2] = exam + question_no   # As our assessments ID named like this
    #print('After: ', evaluate_data_l[0])
    #return evaluate_data_l
    df = pd.DataFrame(evaluate_data_l,columns=['Evaluation ID', 'Student_Enrollment', 'Assessments ID', 'Obtain Marks'] )
    return df
    
```

```python
question_3 = modify(12000, q3, mid_question_3, 'CSE214_Mid_')
#print(question_3[1])
```

```python
question_3.head()
```

```python
# Saving as Excel sheet
question_3.to_excel('evaluation_for_cse214_mid_q3.xlsx')
```
